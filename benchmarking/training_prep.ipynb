## Creating test files

```python
from Bio import SeqIO
```

```python
# Files
alignment_file = "../msa.fasta"  # Alignment file with names like VCAM1_HUMAN
protein_file = "../data/protein-matching-PF00047.fasta"    # Protein database with UniProt IDs
```

```python
# TODO


# Extract identifiers from alignment
alignment_ids = []
with open(alignment_file, "r") as align:
    for line in align:
        if line.startswith(">"):
            alignment_ids.append(line.split("/")[0][1:])  # Extract VCAM1_HUMAN

print(alignment_ids)

# Match with protein FASTA descriptions
matches = []
for record in SeqIO.parse(protein_file, "fasta"):
    print(record)

    for align_id in alignment_ids:
        if align_id in record.description:  # Check if alignment ID is in protein description
            matches.append((align_id, record.id, record.description))

# Output matches
for match in matches:
    print(f"Alignment ID: {match[0]}, Protein ID: {match[1]}, Description: {match[2]}")
  
print(matches)
```
